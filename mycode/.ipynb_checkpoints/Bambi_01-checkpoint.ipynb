```python
import warnings

import arviz as az
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
import bambi as bmb
import pymc as pm
import seaborn  as sns

warnings.simplefilter(action='ignore', category=FutureWarning)

SEED = 1234
az.style.use("arviz-darkgrid")
```

```python
from scipy.stats import zscore

# Example data
size = 200
data = {
    'y': np.random.binomial(1, 0.5, size=size),
    'studentOfColor': np.random.binomial(1, 0.3, size=size),
    'isFirstGeneration': np.random.binomial(1, 0.4, size=size),
    'x1': np.random.normal(0, 1, size=size),
    'x2': np.random.normal(5, 2, size=size)
}

df = pd.DataFrame(data)
df
```

```python
# Scale non-binary predictors to mean 0 and standard deviation 0.5
non_binary_columns = ['x1', 'x2']
df[non_binary_columns] = df[non_binary_columns].apply(lambda x: (x - x.mean()) / (2 * x.std()))

# Convert binary predictors to categorical type if not already
binary_columns = ['studentOfColor', 'isFirstGeneration']
df[binary_columns] = df[binary_columns].astype('category')
df
```

```python
from sklearn.model_selection import train_test_split
df, df_test = train_test_split(df, test_size=0.5, random_state=42)
```

```python
df.info()
```

```python
# Define custom priors
priors = {
    "studentOfColor": bmb.Prior("StudentT", mu=0, sigma=2.5, nu=1),
    "isFirstGeneration": bmb.Prior("StudentT", mu=0, sigma=2.5, nu=1),
    "x1": bmb.Prior("StudentT", mu=0, sigma=2.5, nu=1),
    "x2": bmb.Prior("StudentT", mu=0, sigma=2.5, nu=1)
}

# Define and fit the model using custom priors
model = bmb.Model("y ~ studentOfColor + isFirstGeneration + x1 + x2", df, family="bernoulli", priors=priors)
```

```python
print(model)
model.build()
model.graph()
```

```python
model.plot_priors();
```

```python
# Fit the model
idata = model.fit(
    draws=2000,
    chains=4,
    init="jitter+adapt_diag",
    progressbar=True
)
```

```python
az.plot_trace(idata, compact=False, backend_kwargs={"layout": "constrained"});
```

```python
# Summarize the inference data
az.summary(idata)
```

```python
az.plot_posterior(idata);
```

```python
bmb.interpret.plot_predictions(
    model,
    idata,
    "x1",
    #subplot_kwargs={"main": "x1", "panel": "x2"},
    fig_kwargs={"figsize": (12, 4)},
    legend=True
);
```

### Predict method
The model.predict function in Bambi provides flexibility in generating predictions based on your needs.<br>
The key difference between kind="response_params" and kind="response" lies in the type of predictions they generate. Let's break down the differences and their uses:
1. kind="response_params":
   - This returns draws from the posterior distribution of the likelihood parameters.
   - For a Bernoulli likelihood, it provides the posterior samples of the probability parameter.
   - It is useful when you want to inspect or use the predicted parameters directly (e.g., the predicted probabilities in a logistic regression).
2. kind="response":
    - This returns draws from the posterior predictive distribution, which includes the predicted response variable values.
    - It generates new data points based on the fitted model, incorporating uncertainty from both the posterior parameter estimates and the inherent variability in the data.
    - This is the appropriate choice for posterior predictive checks, as it simulates new observations from the model.

```python
# Predict probabilities for new data
predicted_params = model.predict(idata, kind='response_params', data=df_test, inplace=False)

# Extract the predicted probabilities
pred_probs = predicted_params.posterior['p']

# Calculate the mean predicted probabilities for each new observation
mean_pred_probs = pred_probs.mean(dim=['chain', 'draw']).values

y_pred = np.where(mean_pred_probs >=0.5,1,0)
```

```python
from sklearn.metrics import classification_report
target_names = ['class 0', 'class 1']
print(classification_report(df_test['y'], y_pred, target_names=target_names))
```
